```python
"""
날짜 : 2021/10/18
이름 : 박승필
내용 : 파이썬 Matplotlib 실습하기

Matplotlib 라이브러리
 - 파이썬 데이터 시각화 패키지
 - Matplotlib를 기반으로 하는 seabon 시각화 라이브러리
"""
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 기본 차트

plt.title('plot1')
plt.plot([1, 3, 4, 9, 10, 17])
plt.show()
```

```python
# x축 설정
plt.title('plot2')
plt.plot([10, 20, 30, 40, 50, 60], [1, 3, 4, 9, 10, 17])
plt.show()
```

```python
# 차트 스타일
plt.title('plot3')
plt.plot([10, 20, 30, 40, 50, 60], [1, 3, 4, 9, 10, 17], 'bo--')
plt.grid(True)
plt.show()
```

```python
# 멀티 라인차트
values = np.arange(0, 5, 0.2)
print(values)

plt.title('plot4')
plt.plot(values, values, 'r--')
plt.plot(values, values ** 2, 'g--')
plt.plot(values, values ** 3, 'b--')
plt.show()
```

```python
# 막대 차트
plt.title('plot5')
plt.bar(['apple','banana','orange','grape'], [5, 3, 7, 9])
plt.show()
```

```python
# 히스토그램
dataset1 = np.random.rand(1000) # 0 ~ 1 사이의 임의의 실수 1000개
dataset2 = np.random.randn(1000) # 표준정규분포를 따르는 임의의 실수 1000개

plt.title('Hist1')
plt.hist(dataset1, edgecolor='black')
plt.show()
```

```python
plt.title('Hist2')
plt.hist(dataset2, edgecolor='black')
plt.show()
```

```python
# 산점도
x = np.random.normal(0, 1, 1000) # 표준0 표준편차 1 1000개의 임의의수
y = np.random.normal(0, 1, 1000) 

plt.title('Scatter')
plt.scatter(x,y)
plt.show()
```
